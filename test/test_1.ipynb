```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/EURUSD.ps01_M1_202208190745_202301121620.csv', sep='\t')
df.rename({
    '<DATE>':'Date','<TIME>':'time','<OPEN>':'Open','<HIGH>':'High','<LOW>':'Low','<CLOSE>':'Close','<TICKVOL>':'tickvol','<VOL>':'volume','<SPREAD>':'spread',
},inplace = True,axis = 1)
df['Date'] = df['Date'].apply(lambda x: x.replace('.','-'))
df['Datetime'] = pd.to_datetime(df['Date'] +' '+ df['time'])
df.drop(['Date','time'],axis=1,inplace=True)
df.to_csv('../datasets/eurusd.csv',index=False)
df
```
